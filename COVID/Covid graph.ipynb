```python
import pandas as pd 
time_covid=pd.read_csv('https://api.covid19india.org/csv/latest/case_time_series.csv')
time_covid
```

```python
time_covid['Active Cases']=time_covid['Total Confirmed']-time_covid['Total Recovered']-time_covid['Total Deceased']
```

```python
from matplotlib import pyplot as plt 
Days=[i for i in range(len(time_covid))]
plt.style.use('seaborn-colorblind')
labels=['Active','Recovered','Deceased']
plt.stackplot(Days,time_covid['Active Cases']/1000,time_covid['Total Recovered']/1000,time_covid['Total Deceased']/1000,labels=labels,edgecolor='Black')
plt.title('Stack Plot-COVID India')
plt.xlabel('Days since Breakout',fontsize=14)
plt.ylabel('Number of Patients(in Thousands)',fontsize=14)
plt.legend(loc='upper left')
plt.show()
```
