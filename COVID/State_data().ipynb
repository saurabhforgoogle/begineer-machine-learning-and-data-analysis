```python
State=pd.read_csv('population_india_census2011.csv')
State=State[['State / Union Territory',	'Density']]
State.set_index('State / Union Territory',inplace=True)
State.head(10)
```

```python
State_Today=State_Data[State_Data['Updated On']=='02/07/2020']
State_Today.set_index('State',inplace=True)
```

```python
State_Scatter_Plot=pd.merge(State_Today,State,how='inner',left_index=True,right_index=True)
State_Scatter_Plot.head()
```

```python
State_Scatter_Plot['Updated On']=pd.to_datetime(State_Scatter_Plot['Updated On'],format='%d/%m/%Y')
```

```python
from matplotlib import pyplot as plt 
density=[]
for char in State_Scatter_Plot['Density'].str.split('/'):
    char[0]=char[0].replace(',','')
    density.append(float(char[0]))
Cases=State_Scatter_Plot['Positive']
Tests=State_Scatter_Plot['Total Tested']
plt.scatter(Cases,density,c=Tests,cmap='Greens' ,edgecolors='Black',alpha=0.75)
cbar=plt.colorbar()
cbar.set_label('Accuracy of Cases')
plt.xscale('log')
plt.yscale('log')
plt.title('Plot of total cases vs population Density')
plt.xlabel('Total Cases')
plt.ylabel('Population Density')
plt.show()
```

```python
states=State_Scatter_Plot.sort_values('Positive',ascending=False).iloc[0:3].index.values
Date1=State_Data[State_Data['State']==states[0]]['Updated On']
Case1=State_Data[State_Data['State']==states[0]]['Positive']
Date2=State_Data[State_Data['State']==states[1]]['Updated On']
Case2=State_Data[State_Data['State']==states[1]]['Positive']
Date3=State_Data[State_Data['State']==states[2]]['Updated On']
Case3=State_Data[State_Data['State']==states[2]]['Positive']
```

```python
from datetime import datetime,timedelta
from matplotlib import dates as Ldates


plt.plot_date(Date1,Case1,label=states[0],linestyle='solid',marker=None)
plt.plot_date(Date2,Case2,label=states[1],linestyle='solid',marker=None)
plt.plot_date(Date3,Case3,label=states[2],linestyle='solid',marker=None)
plt.gcf().autofmt_xdate()
plt.legend()
plt.show()
```
