```python
Graph=World[World['location']=='World'][['total_cases','new_cases','total_deaths','location','date']]
```

```python
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
z=[i for i in range(1,184)]
y=[]
for char in Graph['total_cases']:
    y.append(int(char)/1000)
x=[]
for char in Graph['total_deaths']:
    x.append(int(char)/1000)
plt.plot(z,y,label='Total Cases')
plt.plot(z,x,label='Total Death')
plt.xlabel('No of days since Outbreak')
plt.ylabel('Person in Thousand')
plt.legend()
print(plt.style.available)

plt.show()
```
